Hint: Use `df['column_name'].shift(period)`. Check the documentation at https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.shift.htm

```python
data = data[["close"]].copy()
```

```python
lag=3
```

```python
periods = range(lag, 0, -1) # 3, 2, 1
```

```python

data.shift(periods=periods)
```

```python
data = pd.concat([data, data.shift(periods=periods)], axis=1)
```

```python
data.head()
```

```python
data.dropna(axis=0, inplace=True)
```

```python
data.head()
```

```python
X, y = data.drop(columns=["close"]), data["close"]
```

```python


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Input, LSTM, Dense
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
scaler = MinMaxScaler()
train_X = scaler.fit_transform(train_X)
test_X = scaler.transform(test_X)
```

```python
train_X.shape[0]
```

```python
lag
```

```python
train_X.shape[1] // 3
```

```python
(train_X.shape[0], lag, (train_X.shape[1] // lag)) # (samples, timesteps, features)

```

```python
train_X = train_X.reshape(train_X.shape[0], lag, (train_X.shape[1] // lag))
test_X = test_X.reshape(test_X.shape[0], lag, (test_X.shape[1] // lag))
```

```python
model = Sequential([
    Input((lag, (train_X.shape[1] // lag))), # (timesteps, features)
    LSTM(64, activation="relu", return_sequences=True),
    LSTM(32, activation="relu"),
    Dense(1, activation="linear")
])
```

```python
model.summary()
```

```python
model.compile(
    loss="mean_squared_error",
    optimizer="adam",
    metrics=["mean_squared_error"],
)
```

```python
history = model.fit(train_X, train_y, validation_split=0.2, epochs=32, batch_size=8)
```

```python
sns.lineplot(history.history['loss'][1:], label='loss')
sns.lineplot(history.history['val_loss'][1:], label='val_loss')
```

```python
pred_y = model.predict(test_X)
```

```python
r2_score(test_y, pred_y)
```
